## Introducción
Sin duda la pesca artesanal es de vital importancia tanto a nivel mundial como a nivel regional, esto debído a que representa un motor ecómico en las comunidades asentadas en la zona costera, asegurando trabajo y seguridad alimenticia [FAO,2018](http://www.fao.org/3/I9540EN/i9540en.pdf). Es porello que en las discusiones  preliminares  de  Río+20 se introdujo el concepto de economía azul, como un simil de la economía verde al desarrollo de la agricultura. El concepto de economía azul enfoca el aprovechamiento sustentable de los recursos marinos desarrollando las comunidades costeras de una manera integral y diversificada, fomentando actividades alternativas a la pesca que se puedan desarrollar en los litorales [Ivanova,2017](http://200.23.34.40/sociedadyambiente/index.php/sya/article/view/1768/1687)  
<br>
Al 2012 se reportaba que  entre 30% y 40% de la producción pesquera en el mundo correspondía a pesquerías ribereñas o artesanales [Villegas,
2012](https://scholar.google.com/scholar?lookup=0&q=La+pesca+artesanal+como+estrategia+de+sobrevivencia:+el+caso+de+los+pescadores+en+ciudad+del+Carmen,+Campeche&hl=es&as_sdt=0,5),las cuales emplean al 90% de los pescadores del mundo [FAO,2018].  México, siendo el decimosexto país en producción por pesca, no es la excepción, ya que los datos señalan que el 40% de la pesca se realiza de manera artesanal [Ojeda,2012](http://www.redalyc.org/pdf/102/10223040006.pdf). Siendo la zona noroeste del país la más importante en terminos de producción pesquera, la cual está compuesta por los estados de Sonora, Sinaloa, Baja California y Baja California Sur [SAGARPA, 2017](https://www.gob.mx/inapesca/documentos/carta-nacional-pesquera-2017). 

Baja California Sur es un estado con gran diversidad de ecosistemas marinos, la actividad pesquera en la entidad es en su totalidad ribereña, contando con una flota de 3596 embarcaciones y 9700 pescadores. De acuerdo a cifras preliminares, la producción pesquera en el 2017 fue de 118,868 toneladas, representando el 11% de producción nacional, alcanzando el tercer lugar en volumen de producción [SAGARPA, 2012](https://www.gob.mx/inapesca/documentos/carta-nacional-pesquera-2012). 

Por estas características lo diversos autores se plantean la opción de basar en el modelo la economía azul las políticas económicas del estado (teniendo el litoral más extenso de México) [Ivanova,2017](http://200.23.34.40/sociedadyambiente/index.php/sya/article/view/1768/1687).Sin embargo, la pesca es una de las actividades más complejas del sector primario y está influenciada por factores ambientales, económicos, culturales, sociales y políticos. Es por ello que para poder aplicar estas politicas, es necesario un profundo entendimiento profundo de las interacciones de estos factores al momento de tomar decisiones de manejo. 
No obstante que la actividad pesquera de riberaña artesanal no cuenta con estadísticos desagregados, con esto se da pie a que la actividad pesquera se estudie bajo la óptica de datos conglomerados por captura de especie objetivo, sin diferenciar de manera clara los datos estadísticos del recurso (artesanal e industrial) [Magadán-Revelo, et al. 2016](https://www.colpos.mx/wb_pdf/Agroproductividad/2016/AGROPRODUCTIVIDAD_V_2016.pdf#page=44).

Teniendo en consideración lo anterior, se plantea la necesidad de proporcionar herramientas matemáticas que permitan esclarecer los impactos en variaciones de los factores antes mencionades, pero considerando la insertidumbre que genera la falta de datos desagregados de la actividad. Además, estas herramientas deben ponderar la complejidad de los modelos tal como lo menciona [Magnusson et al. 2013](https://onlinelibrary.wiley.com/doi/abs/10.1111/j.1467-2979.2012.00473.x)

Siendo los retos de BCS  similar a la de otras regiones mundiales, se plantea como un caso de estudio ejemplar. Se pueden focalizar dos retos principales, a) el panorama actual de las pesquerías del estado se encuentran desactualizadas, y es necesario estudios que complementen los datos que proporciona la CNP  y b) existe una necesidad de desarrollar modelos que permitan simular la reacción de las pesquerías ante cambios en los factores ambientales, económicos, culturales, sociales y políticos.

Es por lo anterior que el presente trabajo procura anteder estos retos, prorcionando un análisis exploratorio de la base de datos de los avisos de arribo del estado. Además, se plantea que las pesquerías del estado funcionan como un sistema dinámico que reacciona ante cambios en los factores mencionados anteriormente, de este modo se propone un modelo de inteligencia artificial que entienda cada región de pesca como un nodo de información. Dicho nodo supone una cartera de productos pesqueros que se ven presionados por una demanda, los cuales podrían fluctuar su producción al percibir cambios en la vecindad de nodos.

<br> El a el indice de importancia relativo propuesto por [Ojeda et al. 2012](http://www.redalyc.org/pdf/102/10223040006.pdf) se calcula con la forma:
<br><br>
$ \displaystyle IIT_i= {{Valor_i \over Total Valor}  + {Captura_i \over Total Captura} + { Frecuencia_i \over Total Arribo}}*100  $
<br><br>
Además del IIR, se propone un nuevo indice de importancia relativo porcentual el cual se calcula de la siguiente forma:
<br>
<br>
$ \displaystyle IIT_i= {{Valor_i \over Total Valor}  + {Captura_i \over Total Captura} + { Frecuencia_i \over Total Arribo}*100 \over 3 } $

<br>Posteriormente se seleccionaron las pesquerías de mayor importancia relativa, de esta manera se crean los nodos los cuales son representados por diferentes modelos de redes neuronales profundas tales como Long-Short Term Memory (LSTM)[Hochreiter & Schmidhuber  1997](https://doi.org/10.1162/neco.1997.9.8.1735 ), covolutional neuronal network (CNN)[Waseem Rawat & Zenghui Wang 2017](https://doi.org/10.1162/neco_a_00990 ). Estos modelos siguen un entrenamiento supervisado donde los factores son indices de inflación Indice de Precios al Consumidor (IPC), Indice de Confianza del Consumidor (ICC) e Indice de Desempleo Poblacional (IDP) los cuales se descargaron de [INEGI](http://inegi.org.mx/) y las variables ambientales Temperatura Superficial del mar en grados celcius(sst), concentración de clorofila en kg/m^3 (chlor_a)y lluvia en kg/m^2 se obtuvieron de [GIOVANNI](http://giovanni.gsfc.nasa.gov). La variable objetivo de cada modelo es la captura de las pesquerías. Estos modelos fueron optimizados mediante algoritmos de busquedas iterativas de los hiperparámetros y de los parámetros. Para validar los modelos se usa un conjunto de datos testigo el cual fue separado previamente del conjunto de entrenamiento y se calcula el coeficiente de correlación R² de Spearman y la suma cuadrática del error medio de  las simulaciones versus el conjunto testigo.

```python
##Se cargan los paquetes
from pandas import DataFrame
from pandas import concat
from pandas import read_csv
from datetime import datetime
from matplotlib import pyplot
import pandas as pd
import numpy as np
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.stats.multicomp import MultiComparison
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy.stats import shapiro
from tabulate import tabulate
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers.core import Dense, Activation, Dropout
from keras.layers.recurrent import LSTM
from keras.layers.advanced_activations import LeakyReLU
from keras.layers import  MaxPooling1D, Flatten
from keras.callbacks import TensorBoard
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
#importacion de los datos
dataset = read_csv('BDGENERAL.csv', header=0, index_col=0)
#obtener los encabezados
Encabezados=dataset.columns.values.tolist()

#obtener la lista de las oficinas que se tienen dentro de la base de datos
Oficinas=dataset.Oficina.values.tolist()#  extrae la columna de oficinas como una lista
a=pd.Series(Oficinas, dtype="category") #combierte de lista a serie
acat=a.cat.categories #extrae las categorías de la serie
oficinas=pd.Series.tolist(acat) #convierte las categorias en una lista
frecOficina=pd.value_counts(dataset['Oficina'])

#Se obtiene la lista de las pesquerías 
Pesqueria=dataset.Pesqueria.values.tolist()
c=pd.Series(Pesqueria,dtype="category")
acatp=c.cat.categories
pesquerias=pd.Series.tolist(acatp)
frecPesqueria=pd.value_counts(dataset['Pesqueria'])

#Se obtiene la lista de los años
Anos=dataset.Ano.values.tolist()
c=pd.Series(Anos,dtype="category")
acaty=c.cat.categories
años=pd.Series.tolist(acaty)
frecAnos=pd.value_counts(dataset['Ano'])

#Se obtiene la lista de los meses
Mes=dataset.Mes.values.tolist()
c=pd.Series(Mes,dtype="category")
acatm=c.cat.categories
meses=pd.Series.tolist(acatm)
frecMes=pd.value_counts(dataset['Mes'])

#Se crea una lista con los valores filtrados y sumados de las capturas para cada oficina, pesquería, mes y año.
df=['Oficina','Pesq','año','mes','peso','valor','frecuencia'] #se definen los encabezados del arreglo
df1=pd.DataFrame(columns=df) 
```

 Se preparan las series de tiempo, con el algoritmo de la figura 2.

```python
contador=0
for i in range(0,len(oficinas)):
    V=dataset.Oficina==acat[i]
    for j in range(0,len(pesquerias)):
        VV=dataset.Pesqueria==acatp[j]
        for k in range(0,len(años)):
            VVV=dataset.Ano==acaty[k]
            for l in range(0,len(meses)):
                VVVV=dataset.Mes==acatm[l]#genera lista donde ambos son verdaderos
                M=V & VV & VVV & VVVV 
                B=sum(dataset.Peso[M==True])#muestra los registros que cumplen la condición anterior
                C=sum(dataset.Valor[M==True])#muestra la suma del valor para el año en esa oficina
                fr=dataset.Peso[M==True].size
                df1.loc[contador]=[acat[i],acatp[j],acaty[k],acatm[l],B,C,fr]
                contador=contador+1
```

Una vez se tienen las series de tiempo en un DataFrame, se le debe agregar la columna de indice en formato de fechas.

```python
#Se crea una columna con las fechas en formato yyyy-mm-dd y se une al arreglo anterior
dt=['DATE']
dt=pd.DataFrame(columns=dt)
contador=0
for x in range(0,len(df1)):
    z=df1.values[x,2:3]#se extrae la columna de año
    zZ=df1.values[x,3:4]#se extrae la columna de mes
    zzz=datetime(z,zZ,1)# se convierte a formato de fecha
    dt.loc[contador]=[zzz.strftime("%Y-%m-%d")]#se agrega al dataframe
    contador=contador+1

df1 = df1.assign(Date=dt.values)#el data frame creado se une como columna al datafraem que contiene todos los datos
```

Ahora se aplica el algoritmo de clasificación de las oficinas para cada zona, como en la figura 3.

```python
df=['Zona']
ZNc=[]
ZN=pd.DataFrame(ZNc,columns=df)

i=0

for i in range(0,len(df1)):
                if df1.Oficina[i]=='LORETO' or df1.Oficina[i]== 'LA PAZ' or df1.Oficina[i]== 'SANTA ROSALIA' or df1.Oficina[i]=='LOS CABOS':
                    nd1='zn1'
                    ZN.loc[i]=[nd1]
                    i=i+1
                elif  df1.Oficina[i]=='PUERTO ADOLFO LOPEZ MATEOS' or df1.Oficina[i]=='CD. CONSTITUCION'or df1.Oficina[i]=='SAN CARLOS':
                    nd1='zn2'
                    ZN.loc[i]=[nd1]
                    i=i+1
                elif df1.Oficina[i]=='BAHIA TORTUGAS' or df1.Oficina[i]=='BAHIA ASUNCION'or df1.Oficina[i]=='PUNTA ABREOJOS' or df1.Oficina[i]=='ISLA DE CEDROS':
                    nd1='zn3'
                    ZN.loc[i]=[nd1]
                    i=i+1
                elif df1.Oficina[i]=='GUERRERO NEGRO':
                    nd1='zn4'
                    ZN.loc[i]=[nd1]
                    i=i+1
                else:
                    nd1='NaN'
                    ZN.loc[i]=[nd1]
                    i=i+1
```

## Preparación de modelos de IA

 Se carga el archivo creado en el capitulo anterior, donde además se le agregaron las columnas de las variables exógenas. Posteriormente se aplica un "reframe" para recorrer las variables y crear una matriz que se utilizo para crear los conjuntos de entrenamiento y test. Los datos se estandarizan de 0 a 1. 

```python
def series_to_supervised(data, n_in=1, n_out=1, dropnan=True):
    """
    Frame a time series as a supervised learning dataset.
    Arguments:
        data: Sequence of observations as a list or NumPy array.
        n_in: Number of lag observations as input (X).
        n_out: Number of observations as output (y).
        dropnan: Boolean whether or not to drop rows with NaN values.
    Returns:
        Pandas DataFrame of series framed for supervised learning.
    """
    n_vars = 1 if type(data) is list else data.shape[1]
    df = DataFrame(data)
    cols, names = list(), list()
    # input sequence (t-n, ... t-1)
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
    names += [('var%d(t-%d)' % (j+1, i)) for j in range(n_vars)]
    # forecast sequence (t, t+1, ... t+n)
    for i in range(0, n_out):
        cols.append(df.shift(-i))
        if i == 0:
            names += [('var%d(t)' % (j+1)) for j in range(n_vars)]
        else:
            names += [('var%d(t+%d)' % (j+1, i)) for j in range(n_vars)]
    # put it all together
    agg = concat(cols, axis=1)
    agg.columns = names
    # drop rows with NaN values
    if dropnan:
        agg.dropna(inplace=True)
    return agg
```

```python
#importacion de los datos
dataset = read_csv('izn1.csv', header=0, index_col=0)
# ensure all data is float
dataset=dataset.convert_objects(convert_numeric=True)
values = dataset.values
encoder = LabelEncoder()
values[:,1] = encoder.fit_transform(values[:,1])
values = values.astype('float64')
values= pd.DataFrame(values).fillna(0)
scaler = MinMaxScaler(feature_range=(0, 1))
scaled = scaler.fit_transform(values)
reframed = series_to_supervised(scaled, 1, 1)
reframed.drop(reframed.columns[[16,17,18,19,20]], axis=1, inplace=True)
# se dividen los datos en conjunto de entrenamiento y en conjunto de test
values = reframed.values
n_train_hours = len(reframed)-34#aquí divido asígno el numero de datos para test
train = values[:n_train_hours, :]#aquí separo los datos para el entrenamiento
test = values[n_train_hours:, :]#los datos para el test
print(values.shape,train.shape,test.shape)
```

```python
# # se dividen los datos en conjunto de input y en conjunto de output
train_X = train[:, 14:]
train_y = train[:,:15 ]#aquí seleccionas el vector correspondiente a pronóstico
test_X=  test[:, 14:]
test_y = test[:, :15]#idem
# reshape input a 3D [muestra, timesteps, características]
train_X = train_X.reshape((train_X.shape[0], 1, train_X.shape[1]))
test_X = test_X.reshape((test_X.shape[0], 1, test_X.shape[1]))


print(test_X.shape)
```

```python
# diseño del modelo
model = Sequential()
model.add(LSTM(150,return_sequences=True, input_shape=(train_X.shape[1], train_X.shape[2])))
model.add(Dropout(0.2))
model.add(Activation('linear'))
model.add(LeakyReLU(alpha=0.2))
model.add(MaxPooling1D(pool_size=2,padding='same'))
model.add(Dense(15))
model.add(LSTM(150,return_sequences=True,input_shape=(train_X.shape[1], train_X.shape[2])))
model.add(Dropout(0.2))
model.add(Activation('relu'))
model.add(LSTM(50,input_shape=(train_X.shape[1], train_X.shape[2])))
model.add(Dropout(0.6))
model.add(Activation('linear'))
model.add(Dense(15))
model.compile(loss='mae', optimizer='adamax')
tensorboard = TensorBoard(log_dir='./run_test', histogram_freq=1,
                          write_graph=True, write_images=False)
tensorboard.set_model(model) 

history = model.fit(train_X, train_y, epochs=50, batch_size=5,validation_data=(test_X, test_y),verbose=1,shuffle=False,callbacks=[tensorboard])
```
